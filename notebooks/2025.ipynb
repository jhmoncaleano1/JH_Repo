```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer archivo
df = pd.read_excel("../data/2025.xlsx", sheet_name="2025")

# Asegurar columnas numÃ©ricas
df["Gasto"] = pd.to_numeric(df["Gasto"], errors="coerce").fillna(0)
df["Ingreso"] = pd.to_numeric(df["Ingreso"], errors="coerce").fillna(0)

# Agrupar por concepto
acumulados = df.groupby("Concepto")[["Gasto", "Ingreso"]].sum()

# Calcular balance
acumulados["Balance"] = acumulados["Ingreso"] - acumulados["Gasto"]

# Ordenar por Ingreso descendente
acumulados = acumulados.sort_values("Gasto", ascending=False)

# Mostrar resumen en consola
print("ðŸ“Š Cuadro resumen por concepto:\n")
print(acumulados)

# Exportar a Excel
acumulados.to_excel("../data/resumen_por_concepto.xlsx")

# Graficar
acumulados.plot(kind="barh", figsize=(12, 6), color=["red", "green", "blue"], title="Acumulado por Concepto")
plt.ylabel("Total $")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```
